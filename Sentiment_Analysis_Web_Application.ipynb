**Step 3: Create Flask Application**

```python
from flask import Flask, request, render_template_string
from textblob import TextBlob

app = Flask(__name__)

HTML_TEMPLATE = """
<!DOCTYPE html>
<html>
<head>
    <title>Sentiment Analysis App</title>
    <style>
        body{
            font-family: Arial;
            background:#f4f4f4;
            text-align:center;
            padding:40px;
        }

        textarea{
            width:80%;
            height:150px;
            padding:10px;
        }

        button{
            padding:10px 20px;
            background:#007bff;
            color:white;
            border:none;
            cursor:pointer;
        }

        .result{
            margin-top:20px;
            background:white;
            padding:20px;
            border-radius:10px;
        }
    </style>
</head>
<body>

<h1>Sentiment Analysis using TextBlob</h1>

<form method="POST">
    <textarea name="text" placeholder="Enter text here...">{{ user_text if user_text else '' }}</textarea>
    <br><br>
    <button type="submit">Analyze</button>
</form>

{% if sentiment %}
<div class="result">

    <h2>Analysis Result</h2>

    <p>
        <b>Entered Text:</b><br><br>
        "{{ user_text }}"
    </p>

    <hr>

    <p><b>Sentiment:</b> {{ sentiment }}</p>

    <p><b>Polarity:</b> {{ polarity }}</p>

    <p><b>Subjectivity:</b> {{ subjectivity }}</p>

</div>
{% endif %}

</body>
</html>
"""

@app.route("/", methods=["GET", "POST"])
def home():

    sentiment = None
    polarity = None
    subjectivity = None
    user_text = ""

    if request.method == "POST":

        user_text = request.form.get("text", "")

        print("DEBUG USER TEXT =", user_text)

        blob = TextBlob(user_text)

        polarity = blob.sentiment.polarity
        subjectivity = blob.sentiment.subjectivity

        if polarity > 0:
            sentiment = "Positive"
        elif polarity < 0:
            sentiment = "Negative"
        else:
            sentiment = "Neutral"

    return render_template_string(
        HTML_TEMPLATE,
        user_text=user_text,
        sentiment=sentiment,
        polarity=polarity,
        subjectivity=subjectivity
    )
```
